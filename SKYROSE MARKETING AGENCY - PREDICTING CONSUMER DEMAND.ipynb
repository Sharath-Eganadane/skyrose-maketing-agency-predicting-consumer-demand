```python
#Performing Dicky-Fuller Test in order to check if the data is stationary for BeerCraft

import statsmodels.tsa.stattools as sts
print('Result of Dickey-Fuller Test:')
dataoutput_1 = sts.adfuller(Craftbeer['Relative Interest'])
BeerCraft_outputdata = pd.Series(dataoutput_1[0:4], index=['Test Statistics','p-value','# Lags used', '# Observation used'])
for key, value in dataoutput_1[4].items():
   BeerCraft_outputdata['Critical Value (%s)' %key] = value
print(BeerCraft_outputdata)
```

```python
#p value is > 0.05, we infer that the data is not stationary for CraftBeer
```

```python
#Performing Dicky-Fuller Test in order to check if the data is stationary for Whiskey


import statsmodels.tsa.stattools as sts
print('Result of Dickey-Fuller Test:')
dataoutput_2 = sts.adfuller(Whiskey['Relative Interest'])
Whiskey_outputdata = pd.Series(dataoutput_2[0:4], index=['Test Statistics','p-value','# Lags used', '# Observation used'])
for key, value in dataoutput_2[4].items():
    Whiskey_outputdata['Critical Value (%s)' %key] = value
print(Whiskey_outputdata)

if Whiskey_outputdata['p-value'] <= 0.05:
    print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
else:
    print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

# Baseline Methods

# Craftbeer

```python
Craftbeer = Craftbeer.rename(columns = {"Yr-Month": "ds", "Relative Interest": "y"}) 
Craftbeer.tail()
```

```python
#splitting data

Craftbeer_test = Craftbeer[(Craftbeer['ds'] >= '2015-1') & (Craftbeer['ds'] <= '2015-12')]
Craftbeer_train = Craftbeer[(Craftbeer['ds'] <= '2014-12') & (Craftbeer['ds'] >= '2010-1')]
Craftbeer_test = Craftbeer_test.reset_index(drop = True)
Craftbeer_train = Craftbeer_train.reset_index(drop = True)
```

```python
#Plotting Split Data

plt.figure(figsize=(20, 12))
plt.plot(Craftbeer_train['ds'], Craftbeer_train['y'])
plt.xlabel('Yr-Month')
plt.ylabel('Relative Interest')
plt.title('Relative Interest vs Yr-Month')
```

```python
#Train Baseline

Craftbeer_pred = pd.DataFrame(Craftbeer_test['ds'])
Craftbeer_pred = Craftbeer_pred .reset_index(drop = True)
h_max = 1*12+1 
yT = Craftbeer_train['y'][len(Craftbeer_train) - 1]
y1 = Craftbeer_train['y'][0]
T = len(Craftbeer_train)
m = 1 * 12
```

```python
Craftbeer_pred['Mean Baseline'] = Craftbeer_train['y'].mean()
```

```python
Craftbeer_pred['Naive Baseline'] = yT
```

```python
Craftbeer_pred_list = []
for h in range(1, h_max-8): 
  k = math.floor((h - 1)/m)
  index = T + h - m * (k + 1)
  Craftbeer_pred_list.append(Craftbeer['y'][index-1])
Craftbeer_pred['Seasonal Naive Baseline'] = Craftbeer_pred_list
```

```python
Craftbeer_pred
```

```python
Craftbeer_pred.set_index('ds').plot()
```

```python
Craftbeerdrift_list = []
for h in range(1, h_max-8): 
  result = yT + h * ((yT-y1) / T-1)
  Craftbeerdrift_list.append(result)
Craftbeer_pred['Drift Baseline'] = Craftbeerdrift_list
```

```python
figure(figsize=(20,10))
plt.plot(Craftbeer_train['ds'], Craftbeer_train['y'])
plt.plot(Craftbeer_pred['ds'], Craftbeer_pred['Mean Baseline'], label = 'Mean')
plt.plot(Craftbeer_pred['ds'], Craftbeer_pred['Naive Baseline'], label = 'Naive')
plt.plot(Craftbeer_pred['ds'], Craftbeer_pred['Drift Baseline'], label = 'Drift')
plt.plot(Craftbeer_pred['ds'], Craftbeer_pred['Seasonal Naive Baseline'], label = 'Seasonal Naive', linestyle = '--', alpha = 0.3)
plt.xlabel('Date')
plt.ylabel('Relative Interest')
plt.title('Relative Interest vs Time')
plt.legend()
plt.show()
```
